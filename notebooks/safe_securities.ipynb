```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```python
import pandas_datareader.data as web
import datetime
start = datetime.datetime(2016, 12, 1)
```

```python
btc = web.DataReader("BTCUSD=X", 'yahoo',start)
df = pd.DataFrame({'BTC':btc['Open']})
btc.tail(5)
```

```python
gold = web.DataReader("GC=F", 'yahoo',start)
df2 = pd.DataFrame({'GOLD':gold['Open']})
gold.tail(5)
```

```python
chinaInternet = web.DataReader("KWEB", 'yahoo',start)
df3 = pd.DataFrame({'CHINA INTERNET ETF':chinaInternet['Open']})
chinaInternet.tail(5)
```

```python
fidelityMF = web.DataReader("FSELX", 'yahoo',start)
df4 = pd.DataFrame({'FIDELITY MF':fidelityMF['Open']})
fidelityMF.tail(5)
```

```python
columbiaMF = web.DataReader("SCMIX", 'yahoo',start)
df5 = pd.DataFrame({'COLUMBIA MF':columbiaMF['Open']})
columbiaMF.tail(5)
```

```python
globalExUS = web.DataReader("DRW", 'yahoo',start)
df6 = pd.DataFrame({'GLOBAL Ex-US PROP.':globalExUS['Open']})
globalExUS.tail(5)
```

```python
safe_securities = df.merge(pd.DataFrame(df2), left_index=True, right_index=True) \
.merge(pd.DataFrame(df3), left_index=True, right_index=True) \
.merge(pd.DataFrame(df4), left_index=True, right_index=True) \
.merge(pd.DataFrame(df5), left_index=True, right_index=True) \
.merge(pd.DataFrame(df6), left_index=True, right_index=True)
```

```python
safe_securities.dropna(inplace=True)
safe_securities.to_csv('../Data/btc-safe-securities.csv')
safe_securities
```

```python
keywords = ['BTC', 'CHINA INTERNET ETF', 'COLUMBIA MF', 'FIDELITY MF', 'GLOBAL Ex-US PROP.', 'GOLD']
for key in keywords:
    safe_securities[key] = safe_securities[key] - safe_securities[key].shift(5)
safe_securities.dropna(inplace=True)

safe_securities.to_csv('../Data/btc-safe-securities-diff.csv')
safe_securities
```

```python
safe_securities.plot.hexbin(x='BTC',y='GOLD',gridsize=25,cmap='Oranges')
```
