```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```python
import pandas_datareader.data as web
import datetime
start = datetime.datetime(2016, 12, 1)
```

```python
btc = web.DataReader("BTCUSD=X", 'yahoo',start)
df = pd.DataFrame({'BTC':btc['Open']})
btc.tail(5)
```

```python
nikkei = web.DataReader("^N225", 'yahoo',start)
df2 = pd.DataFrame({'NIKKEI':nikkei['Open']})
nikkei.tail(5)
```

```python
euronext = web.DataReader("^N100", 'yahoo',start)
df3 = pd.DataFrame({'EURONEXT':euronext['Open']})
euronext.tail(5)
```

```python
nasdaq = web.DataReader("^IXIC", 'yahoo',start)
df4 = pd.DataFrame({'NASDAQ':nasdaq['Open']})
nasdaq.tail(5)
```

```python
ftse = web.DataReader("^FTSE", 'yahoo',start)
df5 = pd.DataFrame({'FTSE':ftse['Open']})
ftse.tail(5)
```

```python
sse = web.DataReader("000001.SS", 'yahoo',start)
df6 = pd.DataFrame({'SSE':sse['Open']})
sse.tail(5)
```

```python
nyse = web.DataReader("^NYA", 'yahoo',start)
df7 = pd.DataFrame({'NYSE':nyse['Open']})
nyse.tail(5)
```

```python
mktindices = df.merge(pd.DataFrame(df2), left_index=True, right_index=True) \
.merge(pd.DataFrame(df3), left_index=True, right_index=True) \
.merge(pd.DataFrame(df4), left_index=True, right_index=True) \
.merge(pd.DataFrame(df5), left_index=True, right_index=True) \
.merge(pd.DataFrame(df6), left_index=True, right_index=True) \
.merge(pd.DataFrame(df7), left_index=True, right_index=True)
```

```python
mktindices.dropna(inplace=True)
mktindices.to_csv('../Data/btc-world-indices.csv')
mktindices
```

```python
mktindices.plot()
```

```python
keywords = ['BTC','NIKKEI', 'EURONEXT', 'FTSE', 'NASDAQ', 'NYSE', 'SSE']
for key in keywords:
    mktindices[key] = mktindices[key] - mktindices[key].shift(7)
mktindices.dropna(inplace=True)
mktindices.to_csv('../Data/btc-world-indices-diff.csv')
mktindices
```

```python
# Some visualisations for the correlation between NYSE and BTC

mktindices[['BTC','NYSE']].plot(figsize=(15,10))
```

```python
plt.xcorr(mktindices['BTC'], mktindices['NYSE'])
```

```python
from matplotlib.ticker import NullFormatter
nullfmt = NullFormatter()  

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.02

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom_h, width, 0.2]
rect_histy = [left_h, bottom, 0.2, height]
plt.figure(1, figsize=(8, 8))

axScatter = plt.axes(rect_scatter)
axHistx = plt.axes(rect_histx)
axHisty = plt.axes(rect_histy)

axHistx.xaxis.set_major_formatter(nullfmt)
axHisty.yaxis.set_major_formatter(nullfmt)

axScatter.scatter(mktindices['BTC'], mktindices['NYSE'])
```
